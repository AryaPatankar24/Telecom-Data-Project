#5. How long have customers typically stayed with the company (tenure)? What is the distribution of tenure? Are there distinct groups of customers based on tenure (e.g., new, established, long-term)?

```python
df['tenure'].describe()
```

```python
sns.histplot(df['tenure'], bins = 30, kde = True,color = 'Blue')
```

```python
#new - <= 12 months
#established   -  13-48
#longterm - >48 months
```

```python
def categorize_tenure(tenure):
  if tenure <= 12:
    return "New"
  elif 13 <= tenure <= 48:
    return "Established"
  else:
    return "Long-Term"

df['Tenure_group'] = df['tenure']. apply(categorize_tenure)
```

```python
df.head(1)
```

```python
df['Tenure_group'].value_counts()
```

```python
sns.countplot(x = 'Tenure_group', data = df, palette = 'coolwarm')
```

```python
tenure_churn = df.groupby ('Tenure_group')['Churn'].value_counts().unstack()
```

```python
tenure_churn
```

```python
sns.countplot(x = 'Tenure_group', hue = 'Churn', data = df, palette = 'Set1')
```
